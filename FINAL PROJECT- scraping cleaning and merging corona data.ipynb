```python
import pandas as pd 
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from datetime import datetime, timedelta,date
```

```python
def to_data_frame1(info):
    res = pd.DataFrame()
    columns=''
    dataArr=[]
    summary = info.find('div', class_ = "table-responsive field field-name-field-pt-table").find('thead').find('tr')
    
    ### create columns names
    columns=(summary.text)
    columns=columns.splitlines()
    
    ### Data list to Add into dataframe         
    data=info.find('div',class_="table-responsive field field-name-field-pt-table").find('tbody').find_all('tr')
    
    ### fill in the data after arrenge 
    for i in data:
        dataArr.append(i.text.splitlines())

    ### fill in the values in data frame
    for i in range(len(dataArr)-1):
        for j in range(len(columns)):
            res.loc[i,columns[j]]=dataArr[i][j]
    res.to_csv (r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\Situation update worldwide\Situation update worldwide 2020-019.csv', index = False, header=True)
    return res
```

```python
def to_data_frame2(info):
    res = pd.DataFrame()
    columns=['CountryOther','TotalCases','NewCases','TotalDeaths','NewDeaths','TotalRecovered','ActiveCases','SeriousCritical','TotCases-1Mpop','Deaths-1M pop','TotalTests','Tests-1M pop']
    print(len(columns))
    dataArr=[]
###################################################################################################################
### The columns names are very hard to split, so i have written the columns array by spliting the string by my own

    #summary = info.find('div', class_ = "main_table_countries_div").find('thead').find('tr')
  
    ### create columns names
    #columns=(summary.text)
    #columns=columns.splitlines()
####################################################################################################################  
    
    ### Data list to Add into dataframe         
    data=info.find('div',class_="main_table_countries_div").find('tbody').find_all('tr')
    
    ### fill in the data after arrenge 
    for i in data:
        dataArr.append(i.text.splitlines())

    ### fill in the values in data frame
    for i in range(len(dataArr)):
        for j in range(len(columns)):
            res.loc[i,columns[j]]=dataArr[i][j+1]
    
    res.to_csv (r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\coronaData\coronaData-2-019.csv', index = False, header=True)
    return res
```

```python
def to_data_frame3(info):
    res = pd.DataFrame()
    columns=[]
    colAftClean=[]
    dataArr=[]
    dataAftercleaning=[]
    summary = info.find('div', {"id" : 'mw-content-text'}).find("tr")
    
    ### create columns names
    columns=summary.text.splitlines()
    for i in columns:
        if i=='':
            continue
        colAftClean.append(i)
    
    ### pull the Data list to Add into dataframe         
    table_body=info.find('tbody')
    rows = table_body.find_all('tr')
    for row in rows:
        cols=row.find_all('td')
        cols=[x.text.strip() for x in cols]
        dataArr.append(cols)
    for i in range(len(dataArr)):
        if i==0:
            continue
        dataAftercleaning.append(dataArr[i])

    ### fill in the values in data frame
    for i in range(len(dataAftercleaning)):
        for j in range(len(colAftClean)):
            res.loc[i,colAftClean[j]]=dataAftercleaning[i][j]
    
    res.to_csv (r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\ranking_of_health_system\ranking_of_health_systems-2020-019.csv', index = False, header=True)
    return res
```

```python
### input the URLs of sites with interesting data
url1 = 'https://www.ecdc.europa.eu/en/geographical-distribution-2019-ncov-cases'
url2 = 'https://www.worldometers.info/coronavirus/?utm_campaign=homeAdvegas1?'
url3 = 'https://en.wikipedia.org/wiki/World_Health_Organization_ranking_of_health_systems_in_2000#Ranking'
urlsArr=[url1,url2,url3]
count=1

### scrape all data in loop with call functions that get the URLs and do it sperated for eace one
for i in urlsArr:
    driver = webdriver.Chrome(executable_path=r'C:\\webdrivers\\chromedriver.exe')
    if count==1:
        driver.get(i)
    elif count==2:
        driver.get(i)
    elif count==3:
        driver.get(i)
    wait = WebDriverWait(driver, 10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    if count==1:
        Situation=to_data_frame1(soup)
        driver.quit()
    elif count==2:
        coronaData=to_data_frame2(soup)
        driver.quit()
    elif count==3:
        ranking_of_health_systems=to_data_frame3(soup)
        driver.quit()
    count+=1
```

### Clean all data frames 

```python
### Read CSVs files - coronaData ,ranking_of_health_systems-2020 ,Situation update worldwide 2020
coData=pd.read_csv(r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\coronaData\coronaData-2-019.csv')
ranking_of_health_system=pd.read_csv(r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\ranking_of_health_system\ranking_of_health_systems-2020-019.csv')
situation_update_worldwide=pd.read_csv(r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\Situation update worldwide\Situation update worldwide 2020-019.csv')
```

```python
coData=coData.loc[5:]
```

```python
coData.head()
```

```python
### Sort data frame from  A - Z 
ccd=coData.sort_values('CountryOther')
ccd.head()
```

```python
ccd.to_csv(r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\coronaData\ccd18.csv')
```

```python
### Give a new index to DF after sorting from A-Z
ccd.reset_index(inplace=True)
ccd.head()
```

```python
# Delete all the row with NaN in country column
ccd=ccd.dropna(subset=['CountryOther'])
```

```python
# Delete the old index column
ccd=ccd.drop(['index'], axis=1)
```

```python
### Null values check
ccd.isnull().sum()
```

```python
### We see that the missing value are need to be zero, do it with fillna function
ccd=ccd.fillna(0)
```

```python
### Null values check after fill in the missing 
ccd.isnull().sum()
```

```python
### Spliting the columns value by each delimiter 
length=len(ccd)
for col in ccd:
    tmp=col
    ccd[tmp]=ccd[tmp].astype(str).str.replace(',','')
    ccd[tmp]=ccd[tmp].astype(str).str.replace('+','')
    ccd[tmp]=ccd[tmp].astype(str).str.replace(' ','')
    ccd[tmp]=ccd[tmp].astype(str).str.replace('\n','')
    ccd[tmp]=ccd[tmp].astype(str).str.replace("'",'')
    ### Convert string type to int 
    for i in range(length):
        if(ccd[tmp][i]==''):
            ccd[tmp][i]='0'
```

```python
### Filnly we done and the data is clean
ccd.to_csv(r'C:\Users\baraziroi\Desktop\Course_Notebooks\Data science\final project\FINAL_PROJECT\data\Screping_data\coronaData\ccd18.csv')
```

# Merge the files


```python
# Change the Country name column to the same name of coData it will be the key for merging
ranking_of_health_system=ranking_of_health_system.rename(columns={"Country": "CountryOther"})
```

```python
# Change the Country name column to the same name of coData it will be the key 
situation_update_worldwide=situation_update_worldwide.rename(columns={"Places reporting cases": "CountryOther"})
```

```python
# Merging the data frames
merged=pd.merge(ccd,situation_update_worldwide,)
merged=pd.merge(merged,ranking_of_health_system,on='CountryOther')
merged
```
