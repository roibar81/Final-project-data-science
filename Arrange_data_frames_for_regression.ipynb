```python
### drop the unnecessary column
for i in range(len(dfs)):
    dfs[i]=dfs[i].drop(['Unnamed: 0'], axis=1)
```

```python
### Reindexing the DFs
for i in range(len(dfs)):
    dfs[i] = dfs[i].set_index('CountryOther')
```

```python
dfs[16]
```

```python
### Creat data frame with necessary columns
corona_df=pd.DataFrame()
for i in dfs:
    for j in i:
        if j=='TotalRecovered':
            total=i[j]
            corona_df[j]=total
        elif j=='TotalCases':
            total=i[j]
            corona_df[j]=total
        elif j=='TotalDeaths':
            total=i[j]
            corona_df[j]=total
        elif j=='ActiveCases':
            total=i[j]
            corona_df[j]=total
        elif j=='SeriousCritical':
            total=i[j]
            corona_df[j]=total
        elif j=='TotalTests':
            total=i[j]
            corona_df[j]=total
        continue
corona_df
```

```python
### Create confimed cases DF per day from lest few weeks for each country into a data frame
confirmedDf=pd.DataFrame()
for i in range(len(dfs)):
    confirmed=dfs[i]['Confirmed cases in the last 14 days']
    colname=dfs[i]['date'][-1]
    confirmedDf[colname]=confirmed
confirmedDf
```
